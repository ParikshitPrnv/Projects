```python
def cost(x,y,m):
    total=np.zeros(len(y))
    for i in range(len(x)):
        total[i]+=(m*x[i]).sum()
    total=(y-total)**2
    return total.sum()/len(x)
```

```python
def step_descent(x,y,lrate,m):
    mnew=np.zeros(x.shape[1])
    num=len(x)
    for i in range(num):
        X=x[i]
        Y=y[i]
        for j in range(len(m)):
            mnew[j]+=(-2/num)*(Y-sum(m*X))*X[j]
            m[j]=m[j]-lrate*mnew[j]
    new_m = m - lrate*mnew
    return new_m
```

```python
def gradient_descent(x,y,lrate,num_i,m):
    
    for i in range(num_i):
        m=step_descent(x,y,lrate,m)
        print(i,': ',cost(x,y,m))
        print(' m: ',m)
        print()
    return m
```

```python
def run(x,y):
    lrate= 0.000000501
    num_i=100
    m=np.zeros(x.shape[1])
    m=gradient_descent(x,y,lrate,num_i,m)
    return m
```

```python
m=run(X,Y)
```
